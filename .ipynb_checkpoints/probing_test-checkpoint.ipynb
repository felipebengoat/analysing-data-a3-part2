```python
# Import ollama and check that llama3.2:3b is available

import ollama

models = ollama.list()
model_names = [m.model for m in models.models]

print("Models available:", model_names)

if "llama3.2:3b" in model_names:
    print("llama3.2:3b is ready")
else:
    print("Model not found. Run in PowerShell: ollama pull llama3.2:3b")
```

```python
# Model to use
MODEL = "llama3.2:3b"

# System prompt: neutral, no mention of book or series
SYSTEM_PROMPT = """You are a literary assistant. Answer questions about 
the story of One Hundred Years of Solitude clearly and concisely. 
Do not mention whether you are referring to a book or a TV series."""

# 5 probing prompts focused on magical realism and ending/tone
PROMPTS = [
    {
        "id": 1,
        "focus": "magical realism",
        "text": "Describe what happens to Remedios the Beauty. \
How is her departure from the world portrayed?"
    },
    {
        "id": 2,
        "focus": "magical realism",
        "text": "How is the insomnia plague described? \
What are its effects on the people of Macondo?"
    },
    {
        "id": 3,
        "focus": "magical realism",
        "text": "What is the significance of the yellow butterflies \
that appear around Mauricio Babilonia?"
    },
    {
        "id": 4,
        "focus": "ending and tone",
        "text": "How does the story of the Buendía family end? \
What is the overall tone of the ending?"
    },
    {
        "id": 5,
        "focus": "ending and tone",
        "text": "What is the fate of Macondo? \
How is it described in the final moments of the story?"
    },
]

print(f"Model: {MODEL}")
print(f"Number of prompts: {len(PROMPTS)}")
for p in PROMPTS:
    print(f"  [{p['id']}] ({p['focus']}) {p['text'][:60]}...")
```

```python
# Run each prompt and store the responses
# This may take a few minutes on 8GB RAM — be patient

responses = []

for p in PROMPTS:
    print(f"Running prompt {p['id']}/5...")

    result = ollama.chat(
        model=MODEL,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",   "content": p["text"]}
        ]
    )

    answer = result["message"]["content"]

    responses.append({
        "id":       p["id"],
        "focus":   p["focus"],
        "prompt":  p["text"],
        "response": answer
    })

    print(f"Done\n")

print("All prompts completed.")
```

```python
# Display all responses in a readable format

for r in responses:
    print("-" * 60)
    print(f"PROMPT {r['id']} — {r['focus'].upper()}")
    print(f"Q: {r['prompt']}")
    print(f"\nA: {r['response']}")
    print()
```

```python
# Save all prompts and responses to a CSV file

import csv

output_file = "cien_anos_probing_results.csv"

fieldnames = ["id", "focus", "prompt", "response"]

with open(output_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(responses)

print(f" Results saved to {output_file}")
print(f" Rows written: {len(responses)}")
```

```python
# Manual annotation rubric:

ANNOTATIONS = [
    {
        "id": 1,
        "focus": "magical realism",
        "label": "hallucination",  # Remedios becomes a flower/tree: not in either version
        "note": "Invents a transformation into nature. Book: ascends to heaven in sheets."
    },
    {
        "id": 2,
        "focus": "magical realism",
        "label": "book",  # Memory loss, social chaos: faithful to the novel
        "note": "Captures insomnia effects well. Minor error: calls it 'Sleeping Plague'."
    },
    {
        "id": 3,
        "focus": "magical realism",
        "label": "hallucination",  # Butterflies appear after death is wrong, they appear while alive
        "note": "Reverses timeline of butterflies. Invents 'flor de mayo' name."
    },
    {
        "id": 4,
        "focus": "ending and tone",
        "label": "book-only",  # Series S2 not released
        "note": "Tone is correct but confuses generations: ends with José Arcadio not Aureliano."
    },
    {
        "id": 5,
        "focus": "ending and tone",
        "label": "book-only",  # Series S2 not released
        "note": "Captures decay and abandonment correctly. Misses the biblical hurricane."
    },
]

print("Annotation summary:")
print(f"{'ID':<4} {'Focus':<20} {'Label':<15} {'Note'}")
print("-" * 80)
for a in ANNOTATIONS:
    print(f"{a['id']:<4} {a['focus']:<20} {a['label']:<15} {a['note']}")
```

```python
# Visualization: bar chart + annotation table per prompt

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from collections import Counter

colors = {
    "book":          "#1D9E75",
    "series":        "#7F77DD",
    "ambiguous":     "#EF9F27",
    "hallucination": "#D85A30",
    "book-only":     "#888780",
}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(
    "Mini Probing Experiment — One Hundred Years of Solitude\n"
    "Model: llama3.2:3b | Annotation: manual",
    fontsize=11, y=1.02
)

# --- Left: bar chart ---
label_counts = Counter(a["label"] for a in ANNOTATIONS)
bar_labels = list(label_counts.keys())
bar_values = list(label_counts.values())
bar_colors = [colors.get(l, "#888780") for l in bar_labels]

axes[0].bar(bar_labels, bar_values, color=bar_colors,
            edgecolor="white", linewidth=0.5)
axes[0].set_title("Label distribution (all prompts)", fontsize=10)
axes[0].set_ylabel("Number of responses")
axes[0].set_ylim(0, 3)
axes[0].tick_params(axis="x", labelsize=9)

# --- Right: annotation table ---
axes[1].axis("off")

col_labels = ["ID", "Focus", "Label", "Note"]
table_data = [
    [
        f"P{a['id']}",
        a["focus"],
        a["label"],
        a["note"][:55] + "…" if len(a["note"]) > 55 else a["note"]
    ]
    for a in ANNOTATIONS
]

table = axes[1].table(
    cellText=table_data,
    colLabels=col_labels,
    cellLoc="left",
    loc="center",
    colWidths=[0.08, 0.22, 0.22, 0.48]
)
table.auto_set_font_size(False)
table.set_fontsize(8)
table.scale(1, 1.8)

# Color label cells by annotation color
for i, a in enumerate(ANNOTATIONS):
    cell = table[i + 1, 2]
    cell.set_facecolor(colors.get(a["label"], "#888780"))
    cell.set_text_props(color="white", fontweight="bold")

axes[1].set_title("Annotation detail per prompt", fontsize=10)

plt.tight_layout()
plt.savefig("probing_results.png", dpi=150, bbox_inches="tight")
plt.show()
```
