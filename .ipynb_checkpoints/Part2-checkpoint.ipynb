```python
# Inspect the raw file to detect separator and encoding
from pathlib import Path

train_path = Path("data") / "genreLyrics_train.csv"

# Print first 5 raw lines
print("First 5 raw lines")
with open(train_path, encoding="utf-8", errors="replace") as f:
    for i, line in enumerate(f):
        if i >= 5: break
        print(f"Line {i}: {repr(line[:200])}")
```

```python
# Tab-separated file with multiline lyrics
# quoting=3 tells pandas to ignore quote characters entirely
import pandas as pd
from pathlib import Path

df_train = pd.read_csv(
    Path("data") / "genreLyrics_train.csv",
    sep="\t",
    quoting=3,
    engine="python"
)

df_test = pd.read_csv(
    Path("data") / "genreLyrics_test.csv",
    sep="\t",
    quoting=3,
    engine="python"
)

print("Train shape:", df_train.shape)
print("Test shape: ", df_test.shape)
print("\nColumns:", df_train.columns.tolist())
print("\nGenre distribution (train):")
print(df_train["genre"].value_counts())
print("\nSample lyrics (first 100 chars):")
print(df_train["lyrics"].iloc[0][:100])
```

```python
# The lyrics field contains newlines, so we parse manually.
# Each song starts with a line matching: index \t genre \t first_lyric_line
# Subsequent lines belong to the same song until the next tab-separated header.

import pandas as pd
from pathlib import Path

def load_lyrics_tsv(path):
    records = []
    current = None

    with open(path, encoding="utf-8", errors="replace") as f:
        for line in f:
            parts = line.rstrip("\n").split("\t")

            # New song row: has exactly 3 tab-separated fields
            if len(parts) == 3 and parts[0] != "":
                if current is not None:
                    records.append(current)
                current = {
                    "id":     parts[0],
                    "genre":  parts[1],
                    "lyrics": parts[2]
                }
            # Continuation line: belongs to current song's lyrics
            elif current is not None:
                current["lyrics"] += "\n" + line.rstrip("\n")

    if current is not None:
        records.append(current)

    return pd.DataFrame(records)

df_train = load_lyrics_tsv(Path("data") / "genreLyrics_train.csv")
df_test  = load_lyrics_tsv(Path("data") / "genreLyrics_test.csv")

print("Train shape:", df_train.shape)
print("Test shape: ", df_test.shape)
print("\nGenre distribution (train):")
print(df_train["genre"].value_counts())
print("\nSample lyrics (first 150 chars):")
print(df_train["lyrics"].iloc[0][:150])
```

```python
# Keep only rows with valid genres

VALID_GENRES = [
    "Country", "Electronic", "Folk", "Hip-Hop",
    "Indie", "Jazz", "Metal", "Other",
    "Pop", "R&B", "Rock"
]

df_train = df_train[df_train["genre"].isin(VALID_GENRES)].reset_index(drop=True)
df_test  = df_test[df_test["genre"].isin(VALID_GENRES)].reset_index(drop=True)

print("Train shape after cleaning:", df_train.shape)
print("Test shape after cleaning: ", df_test.shape)
print("\nGenre distribution (train):")
print(df_train["genre"].value_counts())
```

```python
# Running 1499 songs would take hours on 8GB RAM.
# We take a stratified sample: up to 10 songs per genre.
# This keeps genre representation balanced and runtime manageable.

# Stratified sample: up to 10 songs per genre from test set

SONGS_PER_GENRE = 10
RANDOM_SEED     = 42

frames = []
for genre, group in df_test.groupby("genre"):
    n = min(SONGS_PER_GENRE, len(group))
    frames.append(group.sample(n=n, random_state=RANDOM_SEED))

df_sample = pd.concat(frames).reset_index(drop=True)

print("Sample size:", len(df_sample))
print("\nSongs per genre:")
print(df_sample["genre"].value_counts().sort_index())
```

```python
# Zero-shot: no examples given, just task description and genre list
import ollama

MODEL = "llama3.2:3b"

GENRES_STR = ", ".join(VALID_GENRES)

def zero_shot_prompt(lyrics):
    return f"""You are a music genre classifier.
Classify the following song lyrics into exactly one of these genres:
{GENRES_STR}

Reply with the genre name only. No explanation, no punctuation.

Lyrics:
{lyrics[:500]}

Genre:"""

def classify_zero_shot(lyrics):
    response = ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": zero_shot_prompt(lyrics)}]
    )
    return response["message"]["content"].strip()

# Test on one example before running all 110
test_lyrics = df_sample["lyrics"].iloc[0]
test_genre  = df_sample["genre"].iloc[0]

prediction = classify_zero_shot(test_lyrics)

print("True genre:     ", test_genre)
print("Predicted genre:", prediction)
```

```python
# Run zero-shot classification on all 110 songs
# This may take 10-20 minutes on 8GB RAM 

predictions_zero = []

for i, row in df_sample.iterrows():
    pred = classify_zero_shot(row["lyrics"])
    predictions_zero.append(pred)

    if (i + 1) % 10 == 0:
        print(f"  {i + 1}/110 done...")

df_sample["pred_zero"] = predictions_zero

print("\nDone. Preview:")
print(df_sample[["genre", "pred_zero"]].head(10))

# Check for unexpected values in predictions
unexpected = [p for p in predictions_zero if p not in VALID_GENRES]
print(f"\nUnexpected predictions: {unexpected}")
```

```python
# Few-shot: 1 example per genre from train set

EXAMPLES_PER_GENRE = 1

few_shot_examples = []
for genre in VALID_GENRES:
    sample = (
        df_train[df_train["genre"] == genre]
        .sample(n=EXAMPLES_PER_GENRE, random_state=RANDOM_SEED)
    )
    for _, row in sample.iterrows():
        few_shot_examples.append({
            "genre":  genre,
            "lyrics": row["lyrics"][:200]
        })

examples_block = ""
for ex in few_shot_examples:
    examples_block += f"Lyrics:\n{ex['lyrics']}\nGenre: {ex['genre']}\n\n"

def clean_prediction(text):
    cleaned = text.strip().rstrip(".,!?:").strip()
    if cleaned in VALID_GENRES:
        return cleaned
    for g in VALID_GENRES:
        if g.lower() == cleaned.lower():
            return g
    return cleaned

def zero_shot_prompt(lyrics):
    return f"""You are a music genre classifier.
Classify the following song lyrics into exactly one of these genres:
{GENRES_STR}

Reply with the genre name only. No explanation, no punctuation.

Lyrics:
{lyrics[:500]}

Genre:"""

def few_shot_prompt(lyrics):
    return f"""You are a music genre classifier.
Classify song lyrics into exactly one of these genres:
{GENRES_STR}

Here are some examples:

{examples_block}Now classify this song. Reply with the genre name only.

Lyrics:
{lyrics[:500]}

Genre:"""

def classify_zero_shot(lyrics):
    response = ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": zero_shot_prompt(lyrics)}]
    )
    return clean_prediction(response["message"]["content"])

def classify_few_shot(lyrics):
    response = ollama.chat(
        model=MODEL,
        messages=[{"role": "user", "content": few_shot_prompt(lyrics)}]
    )
    return clean_prediction(response["message"]["content"])

# Verify
print("True genre:     ", test_genre)
print("Zero-shot pred: ", classify_zero_shot(test_lyrics))
print("Few-shot pred:  ", classify_few_shot(test_lyrics))
```

```python
# Run few-shot classification on all 110 songs

predictions_few = []

for i, row in df_sample.iterrows():
    pred = classify_few_shot(row["lyrics"])
    predictions_few.append(pred)

    if (i + 1) % 10 == 0:
        print(f"  {i + 1}/110 done...")

df_sample["pred_few"] = predictions_few

print("\nDone. Preview:")
print(df_sample[["genre", "pred_zero", "pred_few"]].head(15))

unexpected = [p for p in predictions_few if p not in VALID_GENRES]
print(f"\nUnexpected predictions: {unexpected}")
```

```python
# Calculate Precision, Recall and F1 per genre for both strategies
from sklearn.metrics import precision_recall_fscore_support, f1_score
import pandas as pd

y_true = df_sample["genre"]
y_zero = df_sample["pred_zero"]
y_few  = df_sample["pred_few"]

# Per-genre metrics for zero-shot
p_z, r_z, f_z, _ = precision_recall_fscore_support(
    y_true, y_zero, labels=VALID_GENRES, zero_division=0
)

# Per-genre metrics for few-shot
p_f, r_f, f_f, _ = precision_recall_fscore_support(
    y_true, y_few, labels=VALID_GENRES, zero_division=0
)

# Build comparison dataframe
metrics_df = pd.DataFrame({
    "Genre":   VALID_GENRES,
    "P_zero":  p_z.round(2),
    "R_zero":  r_z.round(2),
    "F1_zero": f_z.round(2),
    "P_few":   p_f.round(2),
    "R_few":   r_f.round(2),
    "F1_few":  f_f.round(2),
})

print("Per-genre metrics:\n")
print(metrics_df.to_string(index=False))

# Overall weighted F1
f1_zero_overall = f1_score(y_true, y_zero, average="weighted", zero_division=0)
f1_few_overall  = f1_score(y_true, y_few,  average="weighted", zero_division=0)

print(f"\nOverall weighted F1 — Zero-shot: {f1_zero_overall:.2f}")
print(f"Overall weighted F1 — Few-shot:  {f1_few_overall:.2f}")
```

```python
# Visualize per-genre F1 comparison: zero-shot vs few-shot
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(VALID_GENRES))
width = 0.35

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(
    "Genre Classification — Zero-shot vs Few-shot\nModel: llama3.2:3b",
    fontsize=11, y=1.02
)

# Left: F1 per genre
axes[0].bar(x - width/2, metrics_df["F1_zero"], width,
            label="Zero-shot", color="#1D9E75", edgecolor="white")
axes[0].bar(x + width/2, metrics_df["F1_few"],  width,
            label="Few-shot",  color="#7F77DD", edgecolor="white")
axes[0].set_xticks(x)
axes[0].set_xticklabels(VALID_GENRES, rotation=45, ha="right", fontsize=9)
axes[0].set_ylabel("F1 Score")
axes[0].set_ylim(0, 1)
axes[0].set_title("F1 Score per genre")
axes[0].legend()
axes[0].axhline(y=0.18, color="#1D9E75", linestyle="--", linewidth=0.8, alpha=0.6)
axes[0].axhline(y=0.19, color="#7F77DD", linestyle="--", linewidth=0.8, alpha=0.6)

# Right: Precision vs Recall scatter for few-shot
axes[1].scatter(metrics_df["R_few"], metrics_df["P_few"],
                color="#7F77DD", s=80, zorder=3)
for i, genre in enumerate(VALID_GENRES):
    axes[1].annotate(genre,
                     (metrics_df["R_few"].iloc[i], metrics_df["P_few"].iloc[i]),
                     textcoords="offset points", xytext=(6, 4), fontsize=8)
axes[1].set_xlabel("Recall")
axes[1].set_ylabel("Precision")
axes[1].set_xlim(-0.05, 1.05)
axes[1].set_ylim(-0.05, 1.05)
axes[1].set_title("Precision vs Recall — Few-shot")
axes[1].plot([0, 1], [0, 1], color="gray", linestyle="--",
             linewidth=0.8, alpha=0.5)

plt.tight_layout()
plt.savefig("classification_results.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# Save results to CSV
metrics_df.to_csv("results/metrics_per_genre.csv", index=False)
df_sample[["genre", "pred_zero", "pred_few"]].to_csv(
    "results/predictions.csv", index=False
)
```
